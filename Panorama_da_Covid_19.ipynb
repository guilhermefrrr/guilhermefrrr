# Panorama da Covid-19

A Covid-19 é a doença provocada pelo novo coronavirus. Em 11 de março de 2020, a Organização Mundial de Saúde (OMS) declarou o surto como uma pandemia, que é uma epidemia que ganha escala global.

Os coronavírus são uma grande família de vírus que causam doenças que variam do resfriado comum a doenças mais graves, como a Síndrome Respiratória do Oriente Médio (MERS-CoV) e a Síndrome Respiratória Aguda Grave (SARS-CoV). A Covid-19 é a doença do coronavírus provocada pela nova cepa descoberta em 2019, que não havia sido identificada anteriormente em seres humanos.

A transmissão se dá pelo contato com gotículas de uma pessoa infectada, seja por meio da tosse, do espirro ou mesmo da fala. 

O Brasil é um dos lugares mais atingidos pela Covid-19 no mundo. Em 2021, a média móvel de mortes diárias pela doença ficou acima de mil durante 191 dias seguidos no país.

Com o objetivo de elevar a consciência situacional a respeito do Covid-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados

As informações obtidas para esse estudo foram retiradas do Our World in Data, organização com milhares de pesquisadores ao redor do mundo que produz pesquisas a partir de dados com o objetivo de combater alguns dos maiores problemas do mundo.

Os dados para algumas variáveis foram coletados pela própria Our World in Data, mas em alguns casos a fonte variou, podendo ser a Universidade Johns Hopkins, organizações internacionais (ONU, OCDE, etc.) ou governos locais.

O leque amplo de variáveis presente nesta base de dados dificulta a uniformidade das informações entre diferentes países. Poucos registram informações sobre os novos pacientes com Covid-19 na UTI em uma semana, por exemplo.

É necessário também ter cuidado com números diários porque o dia da semana pode influenciar para ele ser maior ou menor do que a realidade. Aos fins de semana no Brasil, por exemplo, os registros costumam ficar represados.

### Dicionário de Variáveis

iso_code: Código de três letras para identificar países ou territórios. Esse padrão ISO 3166-1 alpha-3 é definido pela International Organization for Standardization (ISO)

tests_per_case: Média móvel (considerando os últimos sete dias) de testes feitos a cada novo caso confirmado de Covid-19

## Importando os dados

```
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

# importar o dataset COVID-19 para um DataFrame
df = pd.read_csv('owid-covid-data2.csv')
```

A falta de uniformidade nos dados fica evidente aqui. Existem valores ausentes em mais de 98% das entradas para as quatro primeiras variáveis listadas acima. Por outro lado, todas as entradas das variáveis "location", "date" e "iso_code" estão preenchidas

```
df1 = df.loc[df.weekly_icu_admissions > 0].copy()
df1['location'].value_counts()
```

A variável "weekly_icu_admissions" tem 99% de valores ausentes e aqui podemos notar o motivo para isso: apenas 16 países já proporcionaram essa informação em algum momento da pandemia. No caso da Islândia, por exemplo, isso aconteceu apenas em nove dias

```
df.date = pd.to_datetime(df.date)
df.loc[(df.date == '2021-11-07') & (df.continent.notnull())].sort_values("total_deaths", ascending=False).head()
```

Não é possível fazer uma análise exploratória baseado nas datas de uma base de dados se os elementos dessa variável estiverem no tipo "object". Na célula acima foi feita a conversão para um tipo adequado.

Além disso, fizemos dois filtros: selecionamos apenas os países (para não visualizar os números combinados da Europa, por exemplo) e as entradas do dia 7 de novembro de 2021 (data com os dados mais recentes no momento em que fiz essa parte da análise).

Assim, temos como resultado o ranking dos países com mais mortes por Covid-19 até 7 de novembro de 2021.

```
top5_total_deaths = df.loc[(df.date == '2021-11-07') & (df.continent.notnull()), ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[0:5]
top5_total_deaths
```

```
fig, ax = plt.subplots(figsize = (8, 6))
plt.style.use('fivethirtyeight')

ax.bar('location', 'total_deaths', data = top5_total_deaths)

ax.set_title('Mais mortes por Covid-19   |   até 19/10/2021', fontsize = 20, fontweight='bold', fontname = 'Corbel')
ax.set_xlabel('Países', fontsize = 14, fontweight='bold', fontname = 'Corbel')
ax.set_ylabel('Número de mortes', fontsize = 14, fontweight='bold', fontname = 'Corbel')

plt.tight_layout()
```

Com um gráfico, temos uma noção melhor da diferença entre os cinco países com mais mortes por Covid-19 no mundo

```
fig, (ax1, ax2, ax3) = plt.subplots(nrows = 3, ncols = 1, figsize = (8, 16))
plt.style.use('fivethirtyeight')

df.loc[df.location == "World", "total_deaths"].reset_index(drop = True).plot(ax = ax1)
df.loc[df.location == "World", "new_deaths_smoothed"].reset_index(drop = True).plot(ax = ax2)
df.loc[df.location == "Brazil", "new_deaths_smoothed"].reset_index(drop = True).plot(ax = ax3)

ax1.set_title('Covid-19 no mundo  |  até 07/11/2021', fontsize = 20, fontweight='bold', fontname = 'Corbel')
ax1.set_xlabel('Dias de pandemia', fontsize = 14, fontweight='bold', fontname = 'Corbel')
ax1.set_ylabel('Número de mortes (em milhões)', fontsize = 14, fontweight='bold', fontname = 'Corbel')

ax2.set_title('Covid-19 no mundo  |  até 07/11/2021', fontsize = 20, fontweight='bold', fontname = 'Corbel')
ax2.set_xlabel('Dias de pandemia', fontsize = 14, fontweight='bold', fontname = 'Corbel')
ax2.set_ylabel('Média móvel de mortes', fontsize = 14, fontweight='bold', fontname = 'Corbel')

ax3.set_title('Covid-19 no Brasil  |  até 07/11/2021', fontsize = 20, fontweight='bold', fontname = 'Corbel')
ax3.set_xlabel('Dias de pandemia', fontsize = 14, fontweight='bold', fontname = 'Corbel')
ax3.set_ylabel('Média móvel de mortes', fontsize = 14, fontweight='bold', fontname = 'Corbel')

plt.tight_layout()
```

Os três gráficos de linha acima mostram a evolução no número de mortes por Covid-19 em todo o mundo ao longo da pandemia.

O primeiro considera o número acumulado de mortes e dá uma noção do impacto severo que a doença provocou. Através dele, no entanto, não é possível distinguir bem quais foram os momentos mais críticos da pandemia.

O segundo gráfico cumpre essa função, mostrando a média móvel de mortes por Covid-19 em todo o mundo. Apesar de atualmente o número ainda estar elevado (são mais de 6.000 mortes nos últimos sete dias), ele está distante do pico da pandemia e aparenta estar em queda.

O terceiro demonstra como o Brasil seguiu um caminho parecido. Considerando o alto número de mortes registradas, é possível dizer até que a trajetória do país influenciou fortemente os números do mundo como um todo.

Abaixo vamos tentar analisar a relação entre o número de mortes por Covid-19 e o PIB per capita de cada país da base de dados.

Primeiro vamos fazer cópias da tabela original e verificar os rankings nos quesitos "total_deaths", "gdp_per_capita" e "total_deaths_per_million".

```
df_td = df.loc[(df.date == '2021-11-07') & (df.continent.notnull())].sort_values("total_deaths", ascending=False).copy()
df_td = df_td.reset_index(drop = True)
df_td.head()
```

```
df_gdp = df.loc[(df.date == '2021-11-07') & (df.continent.notnull())].sort_values("gdp_per_capita", ascending=False).copy()
df_gdp = df_gdp.reset_index(drop = True)
df_gdp.head()
```

```
df_tdpm = df.loc[(df.date == '2021-11-07') & (df.continent.notnull())].sort_values("total_deaths_per_million", ascending=False).copy()
df_tdpm = df_tdpm.reset_index(drop = True)
df_tdpm.head()
```

```
df2 = df.loc[(df.date == '2021-11-07') & (df.continent.notnull())].sort_values("total_deaths", ascending=False).copy()

fig, ax = plt.subplots(figsize = (8, 8))
plt.style.use('fivethirtyeight')

ax.scatter(x = df2['gdp_per_capita'], y = df2['total_deaths'])

x = 0

for x in range(0, 8):
    plt.text(x = df_td['gdp_per_capita'][x], y = df_td['total_deaths'][x], s = df_td['location'][x])
    
for x in range(0, 2):
    plt.text(x = df_gdp['gdp_per_capita'][x], y = df_gdp['total_deaths'][x], s = df_gdp['location'][x])

ax.set_title('Covid-19 no mundo  |  até 19/10/2021', fontsize = 20, fontweight='bold', fontname = 'Corbel')
ax.set_xlabel('PIB per capita', fontsize = 14, fontweight='bold', fontname = 'Corbel')
ax.set_ylabel('Mortes', fontsize = 14, fontweight='bold', fontname = 'Corbel')

plt.tight_layout()
```

No gráfico acima já podemos ter uma noção de como os países com população mais rica foram menos afetados pela pandemia. Luxemburgo e Catar tiveram poucas mortes por Covid-19 em comparação com Brasil e Índia, por exemplo.

Os Estados Unidos, no entanto, são uma exceção. Apesar do PIB per capita maior que a grande maioria das nações, o país é até o momento o que registrou mais mortes por Covid-19 no mundo.

```
df3 = df.loc[(df.date == '2021-11-07') & (df.continent.notnull())].sort_values("total_deaths_per_million", ascending=False).copy()

fig, ax = plt.subplots(figsize = (8, 8))
plt.style.use('fivethirtyeight')

ax.scatter(x = df3['gdp_per_capita'], y = df3['total_deaths_per_million'])

x = 0

for x in range(0, 8):
    plt.text(x = df_tdpm['gdp_per_capita'][x], y = df_tdpm['total_deaths_per_million'][x], s = df_tdpm['location'][x])
    
for x in range(0, 5):
    plt.text(x = df_gdp['gdp_per_capita'][x], y = df_gdp['total_deaths_per_million'][x], s = df_gdp['location'][x])

ax.set_title('Covid-19 no mundo  |  até 19/10/2021', fontsize = 20, fontweight='bold', fontname = 'Corbel')
ax.set_xlabel('PIB per capita', fontsize = 14, fontweight='bold', fontname = 'Corbel')
ax.set_ylabel('Mortes por milhão', fontsize = 14, fontweight='bold', fontname = 'Corbel')

plt.tight_layout()
```

No gráfico acima, em vez do total de mortes, consideramos as mortes por milhão de pessoas. Esse ajuste serve para evitar distorções no caso de países com uma população muito maior ou menor que os demais.

Ainda assim, a relação entre mortes e riqueza da população fica mais evidente. Os cinco países com mais mortes por milhão se encontram também entre os que não possuem um PIB per capita elevado.

O Brasil, por exemplo, registra pouco menos de 3 mil mortes por milhão (é o sétimo país nesse quesito) e tem PIB per capita parecido com o do Peru.

### Análise Exploratória para o Brasil


Aqui filtramos toda a tabela do Our World in Data para apresentar apenas os dados referentes ao Brasil.

Abaixo podemos ver o registro das informações dos cinco dias mais recentes.

```
df_brasil = df.loc[df.location == "Brazil"].reset_index(drop = True).copy()
df_brasil.tail()
```

```
df_brasil.loc[df_brasil.total_deaths == 1]
```

```
df_brasil['new_deaths_smoothed'].max()
```

```
df_brasil.loc[df_brasil['new_deaths_smoothed'] == 3123.571]
```

"Voltando no tempo", podemos ver que a primeira morte no Brasil foi registrada no dia 17 de março de 2020. A maior média móvel de mortes no país foi de aproximadamente 3.123, registrada no dia 12 de abril de 2021.

```
fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols = 1, figsize = (8, 12))
plt.style.use('fivethirtyeight')

df_brasil['total_deaths'].plot(ax = ax1)
df_brasil['total_cases'].plot(ax = ax1)

ax1.set_title('Covid-19 no Brasil  |  escala linear', fontsize = 20, fontweight='bold', fontname = 'Corbel')

ax1.set_xlabel('Dias de pandemia', fontsize = 14, fontweight='bold', fontname = 'Corbel')
ax1.set_ylabel('em dezenas de milhões', fontsize = 14, fontname = 'Corbel')

ax1.legend(['Mortes', 'Casos'])

ax2.set_title('Covid-19 no Brasil  |  escala log', fontsize = 20, fontweight='bold', fontname = 'Corbel')

df_brasil['total_deaths'].plot(ax = ax2)
df_brasil['total_cases'].plot(ax = ax2)

ax2.set_yscale('log')

ax2.set_xlabel('Dias de pandemia', fontsize = 14, fontweight='bold', fontname = 'Corbel')

plt.tight_layout()
```

No gráfico de cima, em escala linear, a evolução no número de casos é tão grande que faz o número de casos parecer insignificante.

Já o gráfico em escala logarítimica corrige essa distorção, apontando que também houve uma quantidade relevante de mortes por Covid-19 no país ao longo da pandemia.

## Conclusão



Mais de 5 milhões de pessoas já perderam a vida em todo o mundo por conta da Covid-19.

Perto da marca de 100 dias da pandemia houve um pico de cerca de 7 mil na média móvel de mortes pela doença. Esse parecia ser o auge da pandemia, pois na sequência foi notável a queda desse número.

No entanto, entre a marca de 300 e 500 dias de pandemia (ou seja, já no começo de 2021), uma nova onda surgiu e fez a média móvel de mortes em todo o mundo superar a marca de 14 mil.

Nesse momento, já nos distanciamos desse patamar e existe uma tendência de queda mundialmente.

O Brasil é o país com o sétimo maior número de mortes por Covid-19 por milhão de habitantes.

A evolução na média móvel de vidas perdidas seguiu trajetória parecida com o restante do mundo, apresentando seu auge no começo de 2021 e atualmente em queda.

O Brasil é um dos exemplos de como as nações mais atingidas pela pandemia são as que têm a população mais pobre.

Os países com maior PIB per capita do mundo estão longe dos líderes em mortes por milhão de habitantes. Podemos inferir que um maior acesso a cuidados médicos e maiores possibilidades de respeitar o distanciamento social (trabalhando de casa e sem pegar transporte público, por exemplo) ajudaram no controle da pandemia.
